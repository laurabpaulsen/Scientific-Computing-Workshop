```python
df = pd.read_csv("data/height-weight.csv")
```

```python
df.groupby("lives_in_city").mean()
```

```python
df["BMI"] = (df["Weight"] / (df["Height"]**2))*703
df.groupby("lives_in_city").mean()["BMI"]
```

```python
fig, ax = plt.subplots(1, 1, figsize = (6, 4), dpi = 300)

ax.scatter(df[df.lives_in_city == True]["Weight"], df[df.lives_in_city == True]["Height"], marker=".", alpha=0.6, label = "Lives in city")
ax.scatter(df[df.lives_in_city == False]["Weight"], df[df.lives_in_city == False]["Height"], marker=".", alpha=0.6, label = "Does not live in city")

ax.set_ylabel("Weight")
ax.set_xlabel("Height")
ax.legend()

plt.savefig("plot.png")
```
